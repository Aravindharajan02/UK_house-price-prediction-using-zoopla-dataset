```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from catboost import CatBoostRegressor

import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Load and select relevant columns
df = pd.read_csv('/content/Zoopla properties listing information.csv')
columns_to_keep = ['property_size', 'bedrooms', 'bathrooms', 'receptions', 'tenure', 'price_per_size', 'address', 'price']
df = df[columns_to_keep]

# Clean the data
df['tenure'] = df['tenure'].str.extract(r'(Freehold|Leasehold)', expand=False)
df['property_size'] = df['property_size'].str.extract(r'(\d+)', expand=False).astype(float)
df['price_per_size'] = df['price_per_size'].str.replace('£', '', regex=False).str.replace(',', '', regex=False).str.extract(r'(\d+)', expand=False).astype(float)
df['address'] = df['address'].str.extract(r'([A-Z]{1,2}\d{1,2}[A-Z]?)$', expand=False)
df['price'] = df['price'].str.replace('£', '', regex=False).str.replace(',', '', regex=False).str.replace('"', '', regex=False).astype(float)

# Drop rows with missing values
df.dropna(inplace=True)

# One-hot encoding for categorical features
df_encoded = pd.get_dummies(df, columns=['tenure', 'address'], drop_first=True)

# Define features and target
X = df_encoded.drop('price', axis=1)
y = df_encoded['price']

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Models
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Extra Trees": ExtraTreesRegressor(n_estimators=100, random_state=42),
    "ANN (MLP)": MLPRegressor(hidden_layer_sizes=(64, 64), max_iter=1000, random_state=42),
    "KNN": KNeighborsRegressor(n_neighbors=5),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, random_state=42),
    "CatBoost": CatBoostRegressor(verbose=0, random_state=42)
}

# Training and evaluation
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    print(f"\n📊 {name} Performance:")
    print("MAE:", mean_absolute_error(y_test, preds))
    print("MSE:", mean_squared_error(y_test, preds))
    print("RMSE:", np.sqrt(mean_squared_error(y_test, preds)))
    print("R² Score:", r2_score(y_test, preds))
```

```python
# All the setup code remains same as before (loading, cleaning, encoding, scaling, etc.)

# ------- TRAINING CODE START --------
# Keep everything from earlier up to model training...

# Fit models
fitted_models = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    fitted_models[name] = model

# ------- USER INPUT SECTION --------
print("\n🔍 Enter property details to predict the price:")

# Get user input
user_data = {}
user_data['property_size'] = float(input("Enter property size (in sq. ft): "))
user_data['bedrooms'] = int(input("Enter number of bedrooms: "))
user_data['bathrooms'] = int(input("Enter number of bathrooms: "))
user_data['receptions'] = int(input("Enter number of receptions: "))
user_data['price_per_size'] = float(input("Enter price per size (e.g. 2263): "))

# Handle tenure
tenure = input("Enter tenure (Leasehold or Freehold): ").strip().title()
address_code = input("Enter postal code (e.g. WC2R): ").strip().upper()

# One-hot encode manually
# Start from an empty dataframe with same columns as X
user_df = pd.DataFrame([user_data])
for col in df_encoded.columns:
    if 'tenure_' in col:
        user_df[col] = 1 if col == f'tenure_{tenure}' else 0
    elif 'address_' in col:
        user_df[col] = 1 if col == f'address_{address_code}' else 0

# Fill missing columns (if user input didn't match any one-hot)
for col in X.columns:
    if col not in user_df.columns:
        user_df[col] = 0

# Reorder to match model input
user_df = user_df[X.columns]

# Scale user input
user_scaled = scaler.transform(user_df)

# ------- PREDICTIONS --------
print("\n🏡 Predicted property price from each model:\n")
for name, model in fitted_models.items():
    pred = model.predict(user_scaled)[0]
    print(f"{name}: £{pred:,.2f}")
```

```python
# Dictionary to store results
model_scores = []

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    mae = mean_absolute_error(y_test, preds)
    mse = mean_squared_error(y_test, preds)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, preds)

    model_scores.append({
        'Model': name,
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'R² Score': r2
    })

# Convert to DataFrame
results_df = pd.DataFrame(model_scores)
results_df = results_df.sort_values(by='R² Score', ascending=False)

# Display model performance comparison
print("\n📊 Model Performance Comparison:\n")
print(results_df.to_string(index=False))

# Plot (optional)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(data=results_df, x='R² Score', y='Model', palette='viridis')
plt.title("Model Accuracy (R² Score)")
plt.xlabel("R² Score (Higher is better)")
plt.ylabel("Model")
plt.tight_layout()
plt.show()
```
